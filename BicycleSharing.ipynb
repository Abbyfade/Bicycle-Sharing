SPLITTING THE CATEGORICAL DATA

```python
seasons = pd.get_dummies(df_train['season'], prefix='season')
df_train = pd.concat([df_train, seasons], axis=1)
df_train.drop('season', 1, inplace=True)
df_train.head()
```

```python
holidays = pd.get_dummies(df_train['holiday'], prefix='holiday')
df_train = pd.concat([df_train, holidays], axis=1)
df_train.drop('holiday', 1, inplace=True)
df_train.head()
```

```python
workingdays = pd.get_dummies(df_train['workingday'], prefix='workingday')
df_train = pd.concat([df_train, workingdays], axis=1)
df_train.drop('workingday', 1, inplace=True)
df_train.head()
```

```python
weather = pd.get_dummies(df_train['weather'], prefix='weather')
df_train = pd.concat([df_train, weather], axis=1)
df_train.drop('weather', 1, inplace=True)
df_train.head()
```

```python
month = pd.get_dummies(df_train['month'], prefix='month')
df_train = pd.concat([df_train, month], axis=1)
df_train.drop('month', 1, inplace=True)
df_train.head()
```

```python
hours = pd.get_dummies(df_train['hours'], prefix='hours')
df_train = pd.concat([df_train, hours], axis=1)
df_train.drop('hours', 1, inplace=True)
df_train.head()
```

```python
weekday = pd.get_dummies(df_train['weekday'], prefix='weekday')
df_train = pd.concat([df_train, weekday], axis=1)
df_train.drop('weekday', 1, inplace=True)
df_train.head()
```

```python
df_train.drop(['datetime', 'count', 'casual', 'registered', 'atemp'], 1, inplace=True)
```

```python
train_x = df_train
train_x
```

model building

```python
#train_test_split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_x, y.reshape(len(y)), test_size=0.30, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)
```

```python
#Root Mean Squared Logarithmic Error
def rmsle(prediction, actual):
    log1 = np.nan_to_num(np.array([np.log(v + 1) for v in prediction]))
    log2 = np.nan_to_num(np.array([np.log(v + 1) for v in actual]))
    calc = (log1 - log2) ** 2
    return np.sqrt(np.mean(calc))
```

LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
score_lr = rmsle(y_pred_lr, y_test)
print("RMSLE: ", score_lr)
```

RANDOM FOREST

```python
#Instantiating the random forest regressor
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train,y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
score_rf = rmsle(y_pred_rf, y_test)
print("RMSLE: ", score_rf)
```

XGBOOST

```python
from xgboost.sklearn import XGBRegressor
xg = XGBRegressor(objective = 'reg:squarederror', random_state = 42)
xg.fit(X_train, y_train)
```

```python
y_pred_xg = xg.predict(X_test)
```

```python
score_xg = rmsle(y_pred_xg, y_test)
print("RMSLE: ", score_xg)
```

ADABOOST

```python
from sklearn.ensemble import AdaBoostRegressor
adb = AdaBoostRegressor(random_state = 42)
adb.fit(X_train, y_train)
```

```python
y_pred_adb = adb.predict(X_test)
```

```python
score_adb = rmsle(y_pred_adb, y_test)
print("RMSLE: ", score_adb)
```

SVR

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train, y_train)
```

```python
y_pred_svr = svr.predict(X_test)
```

```python
score_svr = rmsle(y_pred_svr, y_test)
print("RMSLE: ", score_svr)
```

BAGGING REGRESSOR

```python
from sklearn.ensemble import BaggingRegressor
br = BaggingRegressor(random_state = 42)
br.fit(X_train, y_train)
```

```python
y_pred_br = br.predict(X_test)
```

```python
score_br = rmsle(y_pred_br, y_test)
print("RMSLE: ", score_br)
```

```python
names = ['Linear Regression', 'Random Forrest', 'XGBoost', 'SVR', 'AdaBoost', 'BaggingRegressor' ]
columns = ['algorithm', 'rmsle']
scores = [score_lr, score_rf, score_xg, score_svr, score_adb, score_br]
alg_vs_score = pd.DataFrame([[x, y] for x, y in zip(names, scores)], columns = columns)
alg_vs_score
```

```python
sns.barplot(data = alg_vs_score, x='algorithm', y='rmsle' )
plt.title('Performance of Different Algorithms')
plt.xticks(rotation='vertical')
```
